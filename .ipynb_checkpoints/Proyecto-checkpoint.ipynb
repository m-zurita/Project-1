```python
#Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
Inflation = pd.read_csv('./Resources/Inflation_Factors.csv')
Inflation_DF = pd.DataFrame(Inflation)
```

```python
# Read the csv
datos = pd.read_csv("./Resources/tmdb_5000_movies.csv")
datos.head(2)
datos_credits = pd.read_csv("./Resources/tmdb_5000_credits.csv")
datos_credits.head(2)
```

```python
# Clean the column genres
generos = []
for i in range(0,len(datos['genres'])):
    try:
        python_dict = json.loads(datos["genres"][i])
        generos.append(python_dict[0]['name'])
    except IndexError:
        generos.append(" ")

genres_df = pd.DataFrame(generos)
genres_df.head(2)
```

```python
# Clean the column production company
production_company = []
for i in range(0,len(datos['production_companies'])):
    try:
        python_dict = json.loads(datos["production_companies"][i])
        production_company.append(python_dict[0]['name'])
    except IndexError:
        production_company.append(" ")
        
production_company_df = pd.DataFrame(production_company)
production_company_df.head(2)
```

```python
# Clean the column country
country =[]
for i in range(0,len(datos['production_countries'])):
    try:
        python_dict = json.loads(datos["production_countries"][i])
        country.append(python_dict[0]['name'])
    except IndexError:
        country.append(" ")
        
country_df = pd.DataFrame(country)
country_df.head(2)
```

```python
# Clean the column cast
cast1 =[]
for i in range(0,len(datos_credits['cast'])):
    try:
        python_dict = json.loads(datos_credits["cast"][i])
        cast1.append(python_dict[0]['name'])
    except IndexError:
        cast1.append(" ")
        
cast1_df = pd.DataFrame(cast1)
cast1_df.head(2)
```

```python
# Clean the column cast
cast2 =[]
for i in range(0,len(datos_credits['cast'])):
    try:
        python_dict = json.loads(datos_credits["cast"][i])
        cast2.append(python_dict[1]['name'])
    except IndexError:
        cast2.append(" ")

cast2_df = pd.DataFrame(cast2)
cast2_df.head(2)
```

```python
# Clean the column cast
cast3 =[]
for i in range(0,len(datos_credits['cast'])):
    try:
        python_dict = json.loads(datos_credits["cast"][i])
        cast3.append(python_dict[2]['name'])
    except IndexError:
        cast3.append(" ")
        
cast3_df = pd.DataFrame(cast3)
cast3_df.head(2)
```

```python
# Clean the column crew and take the information that we need (Director)
director=[]
for j in range(0,len(datos_credits["crew"])):
    try:
        python_dict = json.loads(datos_credits["crew"][j])
        for i in python_dict:
            if i["job"]=="Director":
                director.append(i["name"])
    except IndexError:
        director.append(" ")
        
director_df = pd.DataFrame(director)
director_df.head(2)
```

```python
# Clean the column crew and take the information that we need (Director of Photography)
dp=[]
for j in range(0,len(datos_credits["crew"])):
    try:
        python_dict = json.loads(datos_credits["crew"][j])
        for i in python_dict:
            if i["job"]=="Director of Photography":
                dp.append(i["name"])
    except IndexError:
        dp.append(" ")

dp_df = pd.DataFrame(dp)
dp_df.head(2)
```

```python
# Clean the column crew and take the information that we need (Writer)
writer=[]
for j in range(0,len(datos_credits["crew"])):
    try:
        python_dict = json.loads(datos_credits["crew"][j])
        for i in python_dict:
            if i["job"]=="Writer":
                writer.append(i["name"])
    except IndexError:
        writer.append(" ")

writer_df = pd.DataFrame(writer)
writer_df.head(2)
```

```python
# Clean the column crew and take the information that we need (Producer)
producer=[]
for j in range(0,len(datos_credits["crew"])):
    try:
        python_dict = json.loads(datos_credits["crew"][j])
        for i in python_dict:
            if i["job"]=="Producer":
                producer.append(i["name"])
    except IndexError:
        producer.append(" ")

producer_df = pd.DataFrame(producer)
producer_df.head(2)
```

```python
# Clean the column crew and take the information that we need (Music Composer)
music_composer=[]
for j in range(0,len(datos_credits["crew"])):
    try:
        python_dict = json.loads(datos_credits["crew"][j])
        for i in python_dict:
            if i["job"]=="Original Music Composer":
                music_composer.append(i["name"])
    except IndexError:
        music_composer.append(" ")
        
music_composer_df = pd.DataFrame(music_composer)
music_composer_df.head(2)
```

```python
# Convert both csv to DataFrames
datos_df = pd.DataFrame(datos)
datos_credits_df = pd.DataFrame(datos_credits)
```

```python
main_df = pd.merge(left=datos_df, right=datos_credits_df, how="left")
main_df.head(2)
```

```python
#check if all the columns have the same length
main_df.count()
```

```python
# fill rows that don't have information with "NaN"
Movies_df = main_df.fillna(value="NaN")
Movies_df.count()
```

```python
# Add the columns that we need for the analisis
Movies_df["Genre"] = genres_df
Movies_df["Production Company"] = production_company_df
Movies_df["Country"] = country_df
Movies_df["Leading Role"] = cast1_df
Movies_df["Supporting Role"] = cast2_df
Movies_df["Supporting Role 2"] = cast3_df
Movies_df["Director"] = director_df
Movies_df["Producer"] = producer_df
Movies_df["DP"] = dp_df
Movies_df["Writer"] = writer_df
Movies_df["Music Composer"] = music_composer_df
Movies_df
```

```python
# Delete the columns that we don't need
cut_df = Movies_df.drop(['genres', 'homepage', 'id', 'keywords', 'overview',
              'popularity', 'production_companies', 'status','tagline',
              'original_title', 'production_countries', 'spoken_languages', 'movie_id', 'cast', 'crew'], axis = 1)
```

```python
# Transforming Revenue and Budget into 2017 US Dollars (Part 1: Getting years from the Release Date)
Fechas = []
k = 0
for x in cut_df["release_date"]:
    y = cut_df["release_date"][k].split("-")
    k += 1
    Fechas.append(y)
Fechas_DF = pd.DataFrame(Fechas)
Fechas_DF.columns = ["Year", "Month", "Date"]
```

```python
# (Part 2: Merging inflation data with the original database)
cut_df["Release Year"] = Fechas_DF["Year"]
cut_df2 = cut_df[cut_df['Release Year'] != 'NaN']
pd.options.mode.chained_assignment = None
cut_df2['Release Year'] = cut_df2['Release Year'].astype(int)
Inflation_DF['Year'] = Inflation_DF['Year'].astype(int)
cut_df3 = pd.merge(cut_df2, Inflation_DF, left_on = "Release Year", right_on = "Year")
```

```python
# (Part 3: Generating the inflation adjusted values)
cut_df3["New Budget"] = cut_df3["budget"] * cut_df3["Inflation Factor"]
cut_df3["New Revenue"] = cut_df3["revenue"] * cut_df3["Inflation Factor"]
cut_df3.head()
```

```python
# Rename columns
Renamed_Movies_df = cut_df3.rename(columns = {'New Budget':'Budget',
                                            'original_language':'Language',
                                            'release_date':'Release Date',
                                            'New Revenue':'Revenue',
                                            'runtime':'Runtime',
                                            'title':'Title',
                                            'vote_average':'Rating',
                                            'vote_count':'# of Votes',
                                            })
Renamed_Movies_df.head(2)
```

```python
# Reorganized the columns
organized_movies_df = Renamed_Movies_df[["Title","Production Company", "Genre","Country","Language","Release Date",
                             "Runtime", "Budget", "Revenue", "# of Votes", "Rating","Director","Leading Role",
                             "Supporting Role", "Supporting Role 2", "Producer", "DP", "Writer", "Music Composer"]]
organized_movies_df.head(2)
```

```python
# Apply format to the columns 'Budget', 'Revenue' and '# of Votes'
organized_movies_df['Budget'] = organized_movies_df['Budget'].map('${:,.0f}'.format)
organized_movies_df['Revenue'] = organized_movies_df['Revenue'].map('${:,.0f}'.format)
organized_movies_df['# of Votes'] = organized_movies_df['# of Votes'].map('{:,.0f}'.format)
#organized_movies_df["Budget"].dtype
#organized_movies_df.head(2)
```

# Which actor has made more movies?

```python
len(organized_movies_df['Leading Role'].unique())
len(organized_movies_df['Leading Role'])
```

```python
# Making a dictionary out of Leading Actors
Leading_Actors = organized_movies_df["Leading Role"].unique()
Leading_Actors_Dict = {i : 0 for i in Leading_Actors}
# Checking how many times each actor appeared in a top 5000 movie
for i in organized_movies_df["Leading Role"]:
    Leading_Actors_Dict[i] += 1
# Adding structure to the data
Leading_Actors_DF = pd.DataFrame.from_dict(Leading_Actors_Dict, orient = "index")
Leading_Actors_DF.rename(columns = {0:"Appearance in Movies"}, inplace = True)
Leading_Actors_DF.index.name = "Leading Actors"
Leading_Actors_DF.sort_values(by = "Appearance in Movies", ascending = False, inplace = True)
Leading_Actors_DF.head()
```

```python
# Making a dictionary out of Directors
Directors = organized_movies_df["Director"].unique()
Directors_Dict = {i : 0 for i in Directors}
# Checking how many times each director participated in a top 5000 movie
for i in organized_movies_df["Director"]:
    Directors_Dict[i] += 1
# Adding structure to the data
Directors_DF = pd.DataFrame.from_dict(Directors_Dict, orient = "index")
Directors_DF.rename(columns = {0:"Participation in Movies"}, inplace = True)
Directors_DF.index.name = "Directors"
Directors_DF.sort_values(by = "Participation in Movies", ascending = False, inplace = True)
Directors_DF.head()
```

```python
# Appending Supporting Actors in one list
Supporting_Actors_1 = pd.Series(organized_movies_df["Supporting Role"])
Supporting_Actors_2 = pd.Series(organized_movies_df["Supporting Role 2"])
Supporting_Actors_List = Supporting_Actors_1.append(Supporting_Actors_2)
Supporting_Actors_List
# Making a dictionary out of Supporting Actors
Supporting_Actors = Supporting_Actors_List.unique()
Supporting_Actors_Dict = {i : 0 for i in Supporting_Actors}
# Checking how many times each Supporting Actor participated in a top 5000 movie
for i in organized_movies_df["Supporting Role"]:
    Supporting_Actors_Dict[i] += 1
for i in organized_movies_df["Supporting Role 2"]:
    Supporting_Actors_Dict[i] += 1
# Adding structure to the data
Supporting_Actors_DF = pd.DataFrame.from_dict(Supporting_Actors_Dict, orient = "index")
Supporting_Actors_DF.rename(columns = {0:"Appearance in Movies"}, inplace = True)
Supporting_Actors_DF.index.name = "Supporting Actors"
Supporting_Actors_DF.sort_values(by = "Appearance in Movies", ascending = False, inplace = True)
Supporting_Actors_DF.head(5)
```

```python
# Appending ALL Actors into one List
Actors_1 = pd.Series(organized_movies_df["Leading Role"])
Actors_2 = pd.Series(organized_movies_df["Supporting Role"])
Actors_3 = pd.Series(organized_movies_df["Supporting Role 2"])
Actors_List_1 = Actors_1.append(Actors_2)
Actors_List = Actors_List_1.append(Actors_3)
len(Actors_List)
# Making a dictionary out of ALL Actors
Actors = Actors_List.unique()
Actors_Dict = {i : 0 for i in Actors}
# Checking how many times each Actor participated in a top 5000 movie either on a Leading or Supporting role
for i in organized_movies_df["Leading Role"]:
    Actors_Dict[i] += 1
for i in organized_movies_df["Supporting Role"]:
    Actors_Dict[i] += 1
for i in organized_movies_df["Supporting Role 2"]:
    Actors_Dict[i] += 1
# Adding structure to the data
Actors_DF = pd.DataFrame.from_dict(Actors_Dict, orient = "index")
Actors_DF.rename(columns = {0:"Appearance in Movies"}, inplace = True)
Actors_DF.index.name = "Actors"
Actors_DF.sort_values(by = "Appearance in Movies", ascending = False, inplace = True)
Actors_DF.head()
```
